2) Task-2: Removing Null values from columns 'Age' and 'Fare'.

```python
ans2 = df[['Age', 'Fare']].dropna()
print(ans2.info())
```

3) Task-3: Replacing Null values from columns 'Age' and 'Fare' with mean of their respective columns.

```python
ans3 = df[['Age', 'Fare']].fillna(round(df['Age'].mean()))
print(ans3.info())
ans3.head()
```
